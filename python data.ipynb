```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
def get_game_details(game_urls):

  response = requests.get(game_urls)
  print(response.status_code)
  if response.status_code != 200:
    print("Failed to get game_data!")
    return None


  soup = BeautifulSoup(response.text, "html.parser")

  try:
    rating = soup.find("div", class_="user-rating").text.strip()
  except AttributeError:
    rating = "No info"

  try:
    system_requirements = soup.find("div", class_="system-requirements").text.strip()
  except AttributeError:
    system_requirements = "No info"

  return {"rating": rating, "system_requirements": system_requirements}
```

```python
df_samples = df.head(10)
additional_data = []

for index, row in df_samples.iterrows():
  game_info = get_game_details(row["freetogame_profile_url"])
  if game_info:
    print(game_info)
    additional_data.appand(game_info)

  time.sleep(2)
```
